```python
def plot_hist(variable):
  plt.figure(figsize =(10,5))
  plt.hist(df[variable], bins =50)
  plt.xlabel(variable)
  plt.ylabel("Frequency")
  plt.title("{} distribution with hist" .format(variable))
  plt.show()
```

```python
numericVar = ["Age(yrs)" , "Weight(Kg)" , "Marraige Status(Yrs)"]
for n in numericVar:
  plot_hist(n)
```
